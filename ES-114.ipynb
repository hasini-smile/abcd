```python
n=int(input())
a=0
b=1
l=[]
l.append(a)
if n>1:
    l.append(b)
    i=1
    while i<=(n-2):
        c=a+b
        l.append(c)
        a=b
        b=c
        i+=1
print(l)    
import matplotlib.pyplot as plt
x=list(range(1,(n+1)))
y=l
plt.plot(x,y)
plt.xlabel('Position of element in series')
plt.ylabel('Fibonacci numbers')
plt.title('FIBONACCI SERIES')
plt.show()
```

```python
l=list(map(int, input().split(',')))
x=l
a=min(l)
b=max(l)
y=[]
for i in l:
    h=(i-a)/(b-a)
    y.append(h)
import matplotlib.pyplot as plt
plt.plot(x,y,color='blue')
plt.xlabel('number')
plt.ylabel('normalised number')
plt.title('NORMALISATION')
plt.grid(True)
plt.show()
```
